```python
#This is for 0mM TEMPO (Jack Shared on April 28th) (Faster way)

import numpy as np
import matplotlib.pyplot as plt

TEMPO_0mM_intensity=[10,15,20,25]
TEMPO_0mM_a=[1.807,1.727,1.542,1.483]
TEMPO_0mM_b=[0.496,0.621,0.692,0.771]

slope_a,inter_a=np.polyfit(TEMPO_0mM_intensity,TEMPO_0mM_a,1)
slope_b,inter_b=np.polyfit(TEMPO_0mM_intensity,TEMPO_0mM_b,1)
fig,axes=plt.subplots(1,2,figsize=(12,5))
print(f"Slope: {slope_a:.4f}, Intercept: {inter_a:.4f}")

plt.figure()
axes[0].scatter(TEMPO_0mM_intensity,TEMPO_0mM_a, label='Data Points')
axes[0].plot(TEMPO_0mM_intensity, slope_a*np.array(TEMPO_0mM_intensity) + inter_a, 
         linestyle='--', label=f'Linear Fit: y = {slope_a:.4f}x + {inter_a:.4}')
axes[0].set_xlabel('Intensity')
axes[0].set_ylabel('a')
axes[0].set_title('Absorption vs Intensity_0mM TEMPO')
axes[0].legend()


print(f"Slope: {slope_b:.4f}, Intercept: {inter_b:.4f}")
axes[1].scatter(TEMPO_0mM_intensity,TEMPO_0mM_b, label='Data Points')
axes[1].plot(TEMPO_0mM_intensity, slope_b*np.array(TEMPO_0mM_intensity) + inter_b, 
         linestyle='--', label=f'Linear Fit: y = {slope_b:.4f}x + {inter_b:.4}')
axes[1].set_xlabel('Intensity')
axes[1].set_ylabel('b')
axes[1].set_title('Absorption vs Intensity_0mM TEMPO')
axes[1].legend()
plt.show()


```
